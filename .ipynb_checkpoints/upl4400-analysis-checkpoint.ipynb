# Upper potential limit 4400 mV analysis

```python
# basic imports
%matplotlib inline
from IPython.display import HTML
import analysis
import numpy
import matplotlib.pyplot as plt
import sys
import mpld3
```

```python
%%capture

# Specify the number of trial cells to process.
# For example if we have 15 HE5050 cells then ncells_HE5050 = 15. 
ncells_HE5050_upl4400 = 30
ncells_NMC_upl4400 = 30

# Specifiy the filename as filename_%d where %d indicates that the filenames are listed in order of the number of trials. 
# For example when we create files in the MACCOR they should be created as: filename_1, filename_2 ... filename_31.
# The filename portion should be the same for all trials. 
# We should think of more a systematic filenaming convention moving forward.
fname_HE5050_upl4400 = 'ARGONNE_6_BRL_150206_%d.*'
fname_NMC_upl4400 = 'ARGONNE_6_ZZY523150210-%d.*'

# List of masses corresponding to the filenames.
# TODO: Need to think a more convenient way to enter these values in, perhaps include mass in filename 
mass_HE5050_upl4400 = [8.188,8.28,8.372,8.372,8.372,
                       8.464,8.464,8.464,8.464,8.464,
                       8.464,8.464,8.464,8.464,8.556,
                       8.556,8.556,8.556,8.556,8.556,
                       8.556,8.556,8.556,8.648,8.648,
                       8.648,8.648,8.648,8.464,8.648]

mass_NMC_upl4400 = [13.104,12.897,12.798,12.879,12.69,12.447,13.472,13.338,13.14,13.32,13.266,13.23,12.691,
                    13.059,12.573,12.735,12.33,12.483,12.249,12.492,12.519,12.321,13.149,12.816,13.203,
                    13.176,13.221,12.96,13.059,13.14]

# Calculate average mass.
avg_mass_HE5050_upl4400 = numpy.mean(mass_HE5050_upl4400)
avg_mass_NMC_upl4400 = numpy.mean(mass_NMC_upl4400)

# Create a data object that we can then plot and explore the properties of.
# The format for excuting the command is analysis.get_data(filename,number of cells) where filename and
# number of cells are defined above.
# Name the data object specific to the test, i.e. HE5050-upl4400 for this test or NMC532-upl4500 for another test.
HE5050_upl4400 = analysis.get_data(fname_HE5050_upl4400,ncells_HE5050_upl4400)
NMC_upl4400 = analysis.get_data(fname_NMC_upl4400,ncells_NMC_upl4400)
```

# Below are the raw data plots for cycle metrics: charge capacity (mA h), discharge capacity (mA h), coulombic efficiency, average charge voltage (V), average discharge voltage (V), voltage efficiency, charge energy (W h), discharge energy (W h) and energy efficiency. 
# This will plot the metric versus cycle number for the number of trials you are interested in.

```python
# Enter the range for the cycles of interest.
# For example, if we want to look at cycles 5 to 10 then cyc_range = range(5,11)
# Also, if we want to look at specific cycles that aren't in order cyc_range = [2,5,8,10]
# An error will occur if a cycle range is input that exceeds the number of cycles accomplished in the test thus far. 
HE5050_upl4400_cyc_range = range(1,25)
NMC_upl4400_cyc_range = range(1,25)

# Enter the range for trials of interest.
# For example, if we want to look at trials 8 to 15 then trial_range = range(8,16)
# Also, if we want to look at specific cycles that aren't in order trial_range = [5,13,2,18]
HE5050_upl4400_trial_range = range(1,31)
NMC_upl4400_trial_range = range(1,31)

#Specifiy a color for the group of cells you want to look at for visual clarity
#Good choices: http://colorbrewer2.org/
HE5050_upl4400_clr = '#d95f02'
NMC_upl4400_clr = '#1b9e77'

# This command plots the raw data base metrics for the cells with the indicated cycle range and trial range.
# The format for excuting the command is analysis.plot_raw_basemetrics(testname,cycle range,trial range).
# The testname is the name use 
analysis.plot_raw_basemetrics(HE5050_upl4400,HE5050_upl4400_cyc_range,HE5050_upl4400_trial_range,HE5050_upl4400_clr)
analysis.plot_raw_basemetrics(NMC_upl4400,NMC_upl4400_cyc_range,NMC_upl4400_trial_range,NMC_upl4400_clr)
```

# Question: Should we normalize energy and capacity by active weight of the positive electrode? 
# ... first we should see if there is any correlation with active weight of the measured positive electrode punches and the measured capacity

```python
fig, host = analysis.plot_singlefig_setup()
for i in range(1,ncells_HE5050_upl4400):
    host.plot(mass_HE5050_upl4400[i-1],float(HE5050_upl4400['trial%d' % i]['cycnum10']['tt_drg_cap'])*0.277777778, 
              lw=0, mew=0, ms=8, marker='o', label = 'Cell %d'  % (i), color=HE5050_upl4400_clr, alpha=1.0) 
host.set_xlabel(r'mass (mg)')
host.set_ylabel(r'cycle 10 capacity (mA h)')
host.locator_params(nbins=5)

fig, host = analysis.plot_singlefig_setup()
host.locator_params(nbins=5)
for i in range(1,ncells_NMC_upl4400):
    host.plot(mass_NMC_upl4400[i-1],float(NMC_upl4400['trial%d' % i]['cycnum10']['tt_drg_cap'])*0.277777778, 
              lw=0, mew=0, ms=8, marker='o', label = 'Cell %d'  % (i), color=NMC_upl4400_clr, alpha=1.0) 
host.set_xlabel(r'mass (mg)')
host.set_ylabel(r'cycle 10 capacity (mA h)')
host.locator_params(nbins=5)
```

# The above plots indicate that active mass (as measured by weighing cathode and foil punches) has little to no correlation with capacity. We can just use the average mass or the laminate mass to rescale the plots.

```python
analysis.plot_mscaled_basemetrics(HE5050_upl4400,HE5050_upl4400_cyc_range,HE5050_upl4400_trial_range,HE5050_upl4400_clr,avg_mass_HE5050_upl4400)
analysis.plot_mscaled_basemetrics(NMC_upl4400,NMC_upl4400_cyc_range,NMC_upl4400_trial_range,NMC_upl4400_clr,avg_mass_NMC_upl4400)
```
